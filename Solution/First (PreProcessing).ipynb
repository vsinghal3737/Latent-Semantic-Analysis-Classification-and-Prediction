```python
x = '..\Problem\shuffled-full-set-hashed.csv'
data = pd.read_csv(x, header=None)
df = pd.DataFrame(data)
df.columns = ['output','input']
df.dropna(inplace=True)
```

```python
# WordMapper
mapper = {}
ID = 0
for i in range(len(df)):
    for word in df.iloc[i][1].split(' '):
        if word not in mapper:
            mapper[word] = (word, ID)
            ID += 1
```

```python
# WordSwapperWithMapper
temp = []
x = []
for i in range(len(df)):
    for word in df.iloc[i][1].split(' '):
        temp.append(mapper[word][1])
    df.iloc[i][1] = temp[:]
    temp = []
```

```python
# Output Dic {Output: [Input(s)]}
doc = {}
for i in range(len(df)):
    if df.iloc[i][0] in doc:
        doc[df.iloc[i][0]].append(df.iloc[i][1])
    else:
        doc[df.iloc[i][0]] = [df.iloc[i][1]]
```

```python
# Num of Common Words in each key
CommonWordDoc = {key: {} for key in doc.keys()}
wordCount = {}
for key in CommonWordDoc.keys():
    for subList in doc[key]:
        for numWord in set(subList):
            if numWord in CommonWordDoc[key]:
                CommonWordDoc[key][numWord] += 1
            else:
                CommonWordDoc[key][numWord] = 0
    CommonWordDoc[key] = [
        key,
        list(
            map(
                lambda x: x[0],
                sorted(CommonWordDoc[key].items(), key=lambda x: x[1], reverse=True)[:100]
            )
        )
    ]
```
